```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
y
```

```python
plt.plot(x,y,'b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X-Y graph')
plt.show()
```

```python
plt.subplot(1,3,2)
plt.plot(x,y,'b--')
plt.subplot(1,3,1)
plt.plot(x,y,'r*-')
plt.subplot(1,3,3)
plt.plot(x,y,'g^')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.9,0.99,0.89])
axes.plot(x,y,'b')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('X-Y graph')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.7,0.6,0.9,0.9])
axes2 = fig.add_axes([0.8,0.99,0.4,0.4])
axes1.plot(x,y,'b')
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('X-Y Graph')
axes2.plot(y,x,'r')
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('X-Y Graph')
```

```python
fig,axes=plt.subplots()
axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('X-Y graph');
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('x-y graph')
    fig.tight_layout()
```

```python
fig
```

```python
fig=plt.figure(figsize=(8,4),dpi=100)
fig,axes=plt.subplots(figsize=(8,3))
axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('x-y graph')
```

```python
fig.savefig('filename.png',dpi=200)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x**2')
ax.plot(x,x**3,label='x**3')
ax.legend(loc=2)
```

```python
ax.legend(loc=1)
ax.legend(loc=2)
ax.legend(loc=3)
ax.legend(loc=4)
ax.legend(loc=0)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 6)
plt.plot(x, x + 3, color='b', linewidth=3, linestyle='dashed',
         marker='o', markersize=12,
         markeredgecolor='green', markeredgewidth=4,
         markerfacecolor='red')
plt.plot(x, x + 4, color='r', linewidth=4, linestyle='dashed',
         marker='o', markersize=10,
         markeredgecolor='blue', markeredgewidth=5,
         markerfacecolor='green')
plt.show()
```
